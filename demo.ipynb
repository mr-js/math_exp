```python
sequences = []
for i in range(len(numbers) - 2):
    if numbers[i] == numbers[i + 2]:
        # print((i, numbers[i], numbers[i+1], numbers[i+2]))
        sequences.append((i, numbers[i], numbers[i+1], numbers[i+2]))

if len(sequences) < 2:
    max_interval = 0
    sequence_details = None
else:
    max_interval = 0
    sequence_details = None
    for i in range(len(sequences) - 1):
        current_interval = sequences[i + 1][0] - sequences[i][0]
        if current_interval > max_interval:
            max_interval = current_interval
            sequence_details = (sequences[i], sequences[i + 1])

print("Максимальный интервал между последовательностями:", max_interval)
if sequence_details:
    print("Первая последовательность: =", sequence_details[0][1:])
    print("Индексы первой последовательности:", sequence_details[0][0], "-", sequence_details[0][0] + 2)
    print("Вторая последовательность: =", sequence_details[1][1:])
    print("Индексы второй последовательности:", sequence_details[1][0], "-", sequence_details[1][0] + 2)
else:
    print("Найдено менее двух последовательностей.")
```

```python
import numpy as np

mean = np.mean(numbers)
std_dev = np.std(numbers)

deviations = [(x - mean) / std_dev for x in numbers]

threshold = 1.5
anomalies = [x for x in numbers if abs((x - mean) / std_dev) > threshold]

print(f"Среднее значение {mean}:")
print(f"Стандартное отклонение {std_dev}:")
print(f"Отклонения от среднего значения (в единицах стандартного отклонения): {len(deviations)}",)
print(f"Аномалии (отклонение более чем на {threshold} стандартных отклонений): {anomalies}")
```
